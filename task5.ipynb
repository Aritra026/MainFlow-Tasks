```python
q = ["1. How many people have heart disease and how many people doesnt have heart disease? ",
     "2. People of which sex has most heart disease?",
     "3. People of which sex has which type of chest pain most?",
     "4. People with which chest pain are most prone to have heart disease?"]
q
```

```python
# 1. How many people have heart disease and how many people doesnt have heart disease?
df.target.value_counts()
```

```python
#Plotting bar chart
df.target.value_counts().plot(kind='bar', color=["red", "blue"])
plt.title("Heart Disease Values")
plt.xlabel("1 = Heart Disease, 0 = No Heart Diease")
plt.ylabel("Amount")
```

```python
# pie chart
df.target.value_counts().plot(kind='pie', figsize =(10,8))
plt.legend(['Disease', 'No Disease']);
```

```python
# checking for male and female in the dataset(1 = male, 0 = female)
df.sex.value_counts()
```

```python
#pie chart for sex
df.sex.value_counts().plot(kind='pie', figsize =(10,8))
plt.title('Male Female ratio')
plt.legend(['Male', 'Female'])
```

```python
# 2. Which sex of people has the most heart Disease?
pd.crosstab(df.target, df.sex)
```

```python
# plotting a countplot
sns.countplot(x = 'target', data = df, hue ='sex')
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("0 = No Heart Disease, 1 = Heart Disease")
```

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar', color=['orange','grey','blue','green'])
plt.title("Chest Pain type vs count")
```

```python
# 3. People of which sex has which type of chest pain most?
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind='bar', color=['coral', 'lightskyblue', 'plum', 'khaki'])
plt.title('Type of chest pain for sex')
plt.xlabel('0 = Female , 1 = Male');
```

```python
# 4. People with which chest pain are most prone to have heart disease?
pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x ='cp', data = df, hue = 'target');
```
