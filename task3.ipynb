```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:\MAinFlow Tasks\Task 3\householdtask3.csv')
```

```python
data.head(10)
```

```python
#Scatter plot with age against income
plt.scatter(data['income'],data['age'])
plt.title("Scatter Plot")
plt.xlabel("Income")
plt.ylabel("Age")
plt.show()
```

```python
#Line chart year against expenditure
plt.plot(data['year'])
plt.plot(data['expenditure'])

plt.title("Line Chart")

plt.xlabel('Year')
plt.ylabel('Expenditure')

plt.show()
```

```python
#Bar chart 
plt.bar(data['year'],data['income'])
plt.title("Bar Chart")

plt.xlabel("Year")
plt.ylabel("Income")

plt.show()
```

```python
# Histogram 
plt.hist(data["income"])
plt.title("Histogram")

plt.show()
```
